# seaborn automatically installs below libraries
- numpy
- pandas
- scipy
- matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flower = sns.load_dataset("iris")
flower
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.show()
```

# Adding Title

```python
flower = sns.load_dataset("iris")
flower
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("phoolon ka title")
plt.show()
```

# adding limits

```python
flower = sns.load_dataset("iris")
flower
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("phoolon ka title")

plt.xlim(4.25)
plt.ylim(2.25)
plt.show()
```

# size of figure

```python
flower = sns.load_dataset("iris")
plt.figure(figsize=(4,2))

sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("phoolon ka title")

plt.show()
```
